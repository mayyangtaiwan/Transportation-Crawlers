## Chinahighlights Crawler - www.chinahighlights.com

```python
import requests
from bs4 import BeautifulSoup

head = {"user-agent":"Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_5) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/46.0.2490.80 Safari/537.36"}
payload = { "si":"1",
           "Code_Txt_FZ_1":"",
           "Code_Txt_DZ_1":"",
           "Txt_FZ_1":"Guangzhou",
           "Txt_DZ_1":"Lhasa",
           "stationDate_1":"12/23/2015","Code_Txt_FZ_2":"","Code_Txt_DZ_2":"","Txt_FZ_2":"","Txt_DZ_2":"","stationDate_2":""}    
res = requests.post('http://www.chinahighlights.com/china-trains/search-result.asp', data=payload, headers=head)
soup =  BeautifulSoup(res.text)
searchrsult = soup.select('#searchrsult')[0].findAll('tbody', {'data': True})

l=[]
for i in searchrsult:
    for x in i.findAll('tr', {'data': True}):
        Type = i['data'].split('|')[0]
        Class = x['data'].split('|')[2]
        Price = x['data'].split('|')[5]
        if ('Z' in Type) and ('Standing Ticket' not in Class):
            l.append(Class+','+Price+'\n')

f = open('test.csv', 'w')
for i in l:
    f.write(i)
f.close()

import pandas as pd
df = pd.read_csv('test.csv', header=None)
#print df.sort([0])
MAX = df.groupby([0], sort=False)[1].max()
MIN = df.groupby([0], sort=False)[1].min()
#type(MAX)

concat = pd.concat([MAX, MIN], axis=0, join='inner')
print concat.order(ascending=False)
```
